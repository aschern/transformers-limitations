```python
import pandas as pd

train_data = pd.read_csv('alt_data/train.tsv', sep='\t', header=None)
test_data = pd.read_csv('alt_data/test.tsv', sep='\t', header=None).iloc[1:]
```

```python
import pickle
import numpy as np


with open('alt_data/xlnet_preds', 'rb') as f:
    preds = pickle.load(f).astype(str)
preds[preds == '0'] = 'Task'
preds[preds == '1'] = 'Material'
preds[preds == '2'] = 'Process'
test_data[6] = preds


with open('alt_data/xlnet_preds_logits', 'rb') as f:
    logits = pickle.load(f)
test_data[7] = [' '.join([str(elem) for elem in el]) for el in logits]
```

```python
from sklearn.metrics import f1_score

f1_score(test_data[2].values, test_data[6].values, average='micro')
```

```python
train_labs = {}
for el in set(test_data[0].values) & set(train_data[0].values):
    cor_labels = list(set(train_data[train_data[0] == el][2].values))
    train_labs[el] = cor_labels
```

```python
from copy import deepcopy
from sklearn.metrics import f1_score


map_dict = {'Task': 0, 'Material': 1, 'Process': 2}
labels = ['Task', 'Material', 'Process']
cor_preds = []
for i, row in test_data.iterrows():
    if train_labs.get(row[0]) is not None:
        logs = deepcopy(logits[i - 1])
        for cl in train_labs.get(row[0]):
            logs[map_dict[cl]] += 0.4
        cor_preds.append(labels[np.argmax(logs)])
    else:
        cor_preds.append(row[6])
    #if row[0] in train_labs and row[2] != train_labs.get(row[0])[0]:
    #    print(row[2], row[3], train_labs.get(row[0])[0], logits[i])

f1_score(test_data[2].values, cor_preds, average='micro')
```

## multilabel

```python
def postprocess(x):
    spans_coords = list(zip(x[3].values, x[4].values))
    preds = x[6].values
    logits = [np.array(log.split(), dtype=np.float32) for log in x[7]]
    for i in range(len(preds)):
        log = logits[i]
        
        prev_same = []
        for j in range(i):
            if spans_coords[j][0] == spans_coords[i][0] and spans_coords[j][1] == spans_coords[i][1]:
                prev_same.append(j)
        if len(prev_same) > 0:
            for prediction in preds[prev_same]:
                log[map_dict[prediction]] = 0
                print('done')
        
        logits[i] = log
        preds[i] = labels[np.argmax(log)]
        
    x[6] = preds
    return x
```

```python
f1_score(test_data[2].values, test_data.groupby(5, as_index=False).apply(postprocess)[6], average='micro')
```

## Nesting

```python
from collections import defaultdict

# Task inside Process 5 times p (process|task)

def get_insides(data):  
    insides = defaultdict(dict)
    inv_insides = defaultdict(dict)
    spans_coords = list(zip(data[3].values, data[4].values))
    labels = data[2].values
    article_ids = data[5].values
    for i in range(len(spans_coords)):
        for j in range(i):
            if article_ids[i] == article_ids[j]:
                if spans_coords[i][0] >= spans_coords[j][0] and spans_coords[i][1] <= spans_coords[j][1]:
                    if spans_coords[i][0] != spans_coords[j][0] or spans_coords[i][1] != spans_coords[j][1]:
                        insides[labels[i]][labels[j]] = insides[labels[i]].get(labels[j], 0) + 1
                        inv_insides[labels[j]][labels[i]] = inv_insides[labels[j]].get(labels[i], 0) + 1
                if spans_coords[j][0] >= spans_coords[i][0] and spans_coords[j][1] <= spans_coords[i][1]:
                    if spans_coords[j][0] != spans_coords[i][0] or spans_coords[j][1] != spans_coords[i][1]:
                        insides[labels[j]][labels[i]] = insides[labels[j]].get(labels[i], 0) + 1
                        inv_insides[labels[i]][labels[j]] = inv_insides[labels[i]].get(labels[j], 0) + 1
    return insides, inv_insides
```

```python
from sklearn.utils.extmath import softmax


def correct_preds_for_insides(preds, spans_coords, logits):
    for i in range(len(preds)):
        for j in range(len(preds)):
            if spans_coords[j][0] >= spans_coords[i][0] and spans_coords[j][1] <= spans_coords[i][1]:
                if spans_coords[j][0] != spans_coords[i][0] or spans_coords[j][1] != spans_coords[i][1]:
                    log = softmax([logits[i]])[0]
                    login = softmax([logits[j]])[0]
                    best_prob = 0
                    best_comb = ['Task', 'Task']
                    for cl in labels:
                        for clin in labels:
                            prob = log[map_dict[cl]] * login[map_dict[clin]] * prob_insides[clin][cl] #* prob_inv_insides[cl][clin]
                            if prob > best_prob:
                                preds[i] = cl
                                preds[j] = clin
                                best_prob = prob
    return preds


def postprocess(x):
    spans_coords = list(zip(x[3].values, x[4].values))
    preds = x[6].values
    logits = [np.array(log.split(), dtype=np.float32) for log in x[7]] 
    x[6] = correct_preds_for_insides(x[6].values, spans_coords, logits)
    return x
```
