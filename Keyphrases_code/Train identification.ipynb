```python
import numpy as np


spans = []
start = -1
for i in range(len(pred)):
    if (i == 0 and pred[1].values[i] == 'I') or (pred[1].values[i - 1] == 'O' and pred[1].values[i] == 'I'):
        start = i
    else:
        if pred[1].values[i] == 'O':
            if start != - 1:
                spans.append([start, i])
            start = -1
true_spans = []
start = -1
for i in range(len(pred)):
    if (i == 0 and pred[2].values[i] == 'I') or (pred[2].values[i - 1] == 'O' and pred[2].values[i] == 'I'):
        start = i
    else:
        if pred[1].values[i] == 'O':
            if start != - 1:
                true_spans.append([start, i])
            start = -1
```

```python
def check_closest_start(span, true_spans):
    for true_span in true_spans:
        if true_span[0] < span[0] - 1:
            continue
        if np.abs(true_span[0] - span[0]) <= 1:
            return True
        return False
    
def check_closest_end(span, true_spans):
    for true_span in true_spans:
        if true_span[1] < span[1] - 1:
            continue
        if np.abs(true_span[1] - span[1]) <= 1:
            return True
        return False

near_spans = []
for span in spans:
    if (check_closest_start(span, true_spans) or check_closest_end(span, true_spans)) and span not in true_spans:
        near_spans.append(span)
near_spans
```

```python
def ext_spans(spans):
    new_spans = []
    for span in spans:
        for i in np.arange(5) - 2:
            for j in np.arange(5) - 2:
                if i != 0 or j != 0:
                    new_spans.append([span[0] + i, span[1] + j])
    return set(new_spans)


near_spans = []
ext_true_spans = ext_spans(true_spans)
for span in spans:
    if span in ext_true_spans:
        near_spans.append(span)
near_spans
```
